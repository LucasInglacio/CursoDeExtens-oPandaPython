```python
#filtrar apenas por uma coluna
display(combustiveis_df['Revenda'])
```

```python
#aqui criamos um novo dataframe apenas com as colunas que eu quero...
ca_df = combustiveis_df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda']]
display(ca_df)
```

```python
#exibe a 4a linha. ( as linhas começa do 0) (loc uma pesquisa mais objetiva)
display(ca_df.loc[3])
```

```python
#Exibir da 10a. linha até a 20a linha
display(ca_df.loc[9:19])
```

```python
#Criar uma dataframe gas_df 
#contendo apenas 4 colunas (Revenda, Municipio, Produto, Valor de Venda)
# somente com combustivel sendo GASOLINA e exibir na tela 
#"max" função para descobrir o maior valor 
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)

display(gas_df['Valor de Venda'].max())
```

```python
#Exibir qual é o maior posto  e municipio da gasolina mais cara do país?
display(gas_df[['Revenda', 'Municipio', 'Valor de Venda']].max())
```

```python
#Dataframe.loc[] com múltiplas condições para filtragem
#Quais são os preços, posto que vendem ETANOL na minha cidade BELO HORIZONTE?
# & = ESTRUTURAS 
#inplace= ordena 
etanol_belohorizonte_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'BELO HORIZONTE')]
etanol_belohorizonte_df.sort_values(by='Valor de Venda', inplace=True)
display(etanol_belohorizonte_df) 
```

```python
#Qual a média de preços dos combustiveis GASOLINA e GASOLINA ADITIVADA do bairro MOOCA em SAO PAULO 
#nesse caso vamos usar "combustiveis_df" [v]
#gas_df é so gasoliba comum [x]
# | = OU = 
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') &
                            (combustiveis_df['Municipio'] == 'SAO PAULO') &
                            ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA')),
                            ['Valor de Venda']].mean())

```

```python
#Como mostrar média de valor por Combustiveis? 
#ca_df = pesquisa por colunas
#groupby = ele faz um agrupamento (usado para ter a média, para somar...)
media_por_combustiveis_df = ca_df[['Produto', 'Valor de Venda']].groupby(by='Produto').mean()
display(media_por_combustiveis_df)
```

```python
#Adicionar uma coluna de valor booleano chamda "ATIVO"
#que, por padrão, vai ser true para 
#Todas as linhas
#head = retorna um número especificado de linhas, string a partir do topo. O método head() retorna as primeiras 5 linhas se um número não for especificado. Observação: os nomes das colunas também serão retornados, além das linhas especificadas.

combustiveis_df['Ativo'] = True
print(combustiveis_df.info())
display(combustiveis_df.head())
```
